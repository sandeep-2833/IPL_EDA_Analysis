```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
matches_path = 'matches.csv'
deliveries_path = 'deliveries.csv'

matches_df = pd.read_csv(matches_path)
deliveries_df = pd.read_csv(deliveries_path)

# Styling for plots
sns.set(style="whitegrid")

# Function to analyze matches played per season
def matches_per_season(matches_df):
    data = matches_df['season'].value_counts().sort_index()
    plt.figure(figsize=(12, 6))
    sns.barplot(x=data.index, y=data.values, palette="viridis")
    plt.title("Count of Matches Played in Each Season", fontsize=16)
    plt.xlabel("Season", fontsize=12)
    plt.ylabel("Matches", fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to analyze runs scored per season
def runs_per_season(deliveries_df, matches_df):
    merged_df = deliveries_df.merge(matches_df[['id', 'season']], left_on='match_id', right_on='id', how='left')
    data = merged_df.groupby('season')['total_runs'].sum()
    plt.figure(figsize=(12, 6))
    sns.barplot(x=data.index, y=data.values, palette="magma")
    plt.title("Total Runs Scored in Each Season", fontsize=16)
    plt.xlabel("Season", fontsize=12)
    plt.ylabel("Total Runs", fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to analyze runs per match in each season
def runs_per_match_per_season(matches_df):
    matches_per_season = matches_df['season'].value_counts().sort_index()
    total_runs_per_season = matches_df.groupby('season')['result_margin'].sum()
    data = total_runs_per_season / matches_per_season
    plt.figure(figsize=(12, 6))
    sns.lineplot(x=data.index, y=data.values, marker='o', color='blue')
    plt.title("Runs Scored Per Match in Each Season", fontsize=16)
    plt.xlabel("Season", fontsize=12)
    plt.ylabel("Runs per Match", fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to analyze most frequent umpires
def most_frequent_umpires(matches_df):
    umpires = pd.concat([matches_df['umpire1'], matches_df['umpire2']])
    data = umpires.value_counts().head(10)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=data.values, y=data.index, palette="coolwarm")
    plt.title("Top 10 Most Frequent Umpires", fontsize=16)
    plt.xlabel("Number of Matches", fontsize=12)
    plt.ylabel("Umpire", fontsize=12)
    plt.tight_layout()
    plt.show()

# Function to analyze teams winning the most tosses
def most_tosses_won(matches_df):
    data = matches_df['toss_winner'].value_counts()
    plt.figure(figsize=(12, 6))
    sns.barplot(x=data.values, y=data.index, palette="Blues_r")
    plt.title("Teams Winning the Most Tosses", fontsize=16)
    plt.xlabel("Tosses Won", fontsize=12)
    plt.ylabel("Team", fontsize=12)
    plt.tight_layout()
    plt.show()

# Function to analyze toss decisions
def toss_decisions(matches_df):
    data = matches_df['toss_decision'].value_counts()
    plt.figure(figsize=(6, 6))
    data.plot.pie(autopct='%1.1f%%', colors=['#66b3ff','#99ff99'], startangle=90)
    plt.title("Toss Decisions (Bat or Field)", fontsize=16)
    plt.ylabel('')
    plt.tight_layout()
    plt.show()

# Function to analyze toss decision trends across seasons
def toss_decision_trend(matches_df):
    data = matches_df.groupby('season')['toss_decision'].value_counts(normalize=True).unstack()
    data.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='coolwarm', alpha=0.85)
    plt.title("Toss Decision Trends Across Seasons", fontsize=16)
    plt.xlabel("Season", fontsize=12)
    plt.ylabel("Proportion", fontsize=12)
    plt.legend(title="Decision", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# Function to analyze whether winning the toss implies winning the match
def toss_vs_match_winner(matches_df):
    matches_df['toss_win_match_win'] = matches_df['toss_winner'] == matches_df['winner']
    data = matches_df['toss_win_match_win'].value_counts(normalize=True)
    data.index = ['Lost', 'Won']
    plt.figure(figsize=(6, 6))
    data.plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=90)
    plt.title("Does Winning the Toss Imply Winning the Match?", fontsize=16)
    plt.ylabel('')
    plt.tight_layout()
    plt.show()

# Function to analyze most wins while chasing
def chasing_team_wins(matches_df):
    chasing_wins = matches_df[matches_df['result'] == 'wickets']
    data = chasing_wins['winner'].value_counts().head(10)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=data.values, y=data.index, palette="Greens_r")
    plt.title("Top 10 Teams Winning Most While Chasing", fontsize=16)
    plt.xlabel("Wins", fontsize=12)
    plt.ylabel("Teams", fontsize=12)
    plt.tight_layout()
    plt.show()

def teams_won_tournament(matches_df):
    winners = matches_df.dropna(subset=['winner']).groupby('season')['winner'].last()
    plt.figure(figsize=(12, 6))
    sns.countplot(y=winners, order=winners.value_counts().index, palette="coolwarm")
    plt.title("Teams That Won the Tournament", fontsize=16)
    plt.xlabel("Count", fontsize=12)
    plt.ylabel("Teams", fontsize=12)
    plt.tight_layout()
    plt.show()

def team_with_most_matches(matches_df):
    teams = pd.concat([matches_df['team1'], matches_df['team2']])
    matches_count = teams.value_counts()
    plt.figure(figsize=(12, 6))
    sns.barplot(x=matches_count.values, y=matches_count.index, palette="viridis")
    plt.title("Teams That Played the Most Matches", fontsize=16)
    plt.xlabel("Matches Played", fontsize=12)
    plt.ylabel("Teams", fontsize=12)
    plt.tight_layout()
    plt.show()

def team_with_most_wins(matches_df):
    wins = matches_df['winner'].value_counts()
    plt.figure(figsize=(12, 6))
    sns.barplot(x=wins.values, y=wins.index, palette="magma")
    plt.title("Teams That Won the Most Matches", fontsize=16)
    plt.xlabel("Wins", fontsize=12)
    plt.ylabel("Teams", fontsize=12)
    plt.tight_layout()
    plt.show()

def team_with_highest_win_percentage(matches_df):
    teams = pd.concat([matches_df['team1'], matches_df['team2']]).value_counts()
    wins = matches_df['winner'].value_counts()
    win_percentage = (wins / teams * 100).sort_values(ascending=False)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=win_percentage.values, y=win_percentage.index, palette="Blues_r")
    plt.title("Teams with Highest Winning Percentage", fontsize=16)
    plt.xlabel("Win Percentage", fontsize=12)
    plt.ylabel("Teams", fontsize=12)
    plt.tight_layout()
    plt.show()

def lucky_venue_for_team(matches_df):
    venue_team_wins = matches_df.groupby(['venue', 'winner']).size().reset_index(name='wins')
    top_venues = venue_team_wins.loc[venue_team_wins.groupby('winner')['wins'].idxmax()]
    plt.figure(figsize=(12, 6))
    sns.barplot(x='wins', y='winner', hue='venue', data=top_venues, dodge=False, palette="Set2")
    plt.title("Lucky Venue for Each Team", fontsize=16)
    plt.xlabel("Wins", fontsize=12)
    plt.ylabel("Teams", fontsize=12)
    plt.legend(title="Venue", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

def comparison_between_two_teams(deliveries_df, team1, team2):
    head_to_head = deliveries_df[(deliveries_df['batting_team'].isin([team1, team2])) & 
                                 (deliveries_df['bowling_team'].isin([team1, team2]))]
    head_to_head_runs = head_to_head.groupby(['batting_team'])['total_runs'].sum()
    plt.figure(figsize=(12, 6))
    sns.barplot(x=head_to_head_runs.values, y=head_to_head_runs.index, palette="coolwarm")
    plt.title(f"Comparison Between {team1} and {team2}", fontsize=16)
    plt.xlabel("Total Runs", fontsize=12)
    plt.ylabel("Teams", fontsize=12)
    plt.tight_layout()
    plt.show()

def batsman_analysis(deliveries_df, batsman):
    batsman_data = deliveries_df[deliveries_df['batter'] == batsman]
    runs = batsman_data['batsman_runs'].sum()
    balls = batsman_data.shape[0]
    strike_rate = (runs / balls) * 100
    print(f"Analysis for {batsman}:")
    print(f"Total Runs: {runs}, Balls Faced: {balls}, Strike Rate: {strike_rate:.2f}")

    batsman_performance = batsman_data.groupby(['match_id'])['batsman_runs'].sum()
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=batsman_performance, marker="o", color='purple')
    plt.title(f"Performance Trend of {batsman}", fontsize=16)
    plt.xlabel("Match ID", fontsize=12)
    plt.ylabel("Runs Scored", fontsize=12)
    plt.tight_layout()
    plt.show()


# Define analysis functions
def teams_conceding_200_plus(deliveries_df):
    high_scores = deliveries_df.groupby(['match_id', 'bowling_team'])['total_runs'].sum()
    high_scores = high_scores[high_scores > 200].reset_index()
    most_conceded = high_scores['bowling_team'].value_counts()
    plt.figure(figsize=(12, 6))
    sns.barplot(x=most_conceded.values, y=most_conceded.index, palette="coolwarm")
    plt.title("Teams Conceding 200+ Scores the Most", fontsize=16)
    plt.xlabel("Count", fontsize=12)
    plt.ylabel("Teams", fontsize=12)
    plt.tight_layout()
    plt.show()

def highest_team_score(deliveries_df):
    scores = deliveries_df.groupby(['match_id', 'batting_team'])['total_runs'].sum()
    highest_score = scores.max()
    team = scores.idxmax()[1]
    print(f"The highest score is {highest_score}, scored by {team}.")

def biggest_win_by_runs(matches_df):
    max_margin = matches_df[matches_df['result'] == 'runs']['result_margin'].max()
    team = matches_df[matches_df['result_margin'] == max_margin]['winner'].iloc[0]
    print(f"The biggest win by runs is {max_margin}, achieved by {team}.")

def most_balls_faced_batsmen(deliveries_df):
    balls_faced = deliveries_df.groupby('batter').size().sort_values(ascending=False).head(10)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=balls_faced.values, y=balls_faced.index, palette="viridis")
    plt.title("Batsmen with Most Balls Faced", fontsize=16)
    plt.xlabel("Balls Faced", fontsize=12)
    plt.ylabel("Batsmen", fontsize=12)
    plt.tight_layout()
    plt.show()

def leading_run_scorers(deliveries_df):
    runs = deliveries_df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=runs.values, y=runs.index, palette="magma")
    plt.title("Leading Run Scorers of All Time", fontsize=16)
    plt.xlabel("Total Runs", fontsize=12)
    plt.ylabel("Batsmen", fontsize=12)
    plt.tight_layout()
    plt.show()

def most_boundaries(deliveries_df, boundary_type='4'):
    boundary_counts = deliveries_df[deliveries_df['batsman_runs'] == int(boundary_type)].groupby('batter').size().sort_values(ascending=False).head(10)
    title = f"Batsmen with Most Number of {boundary_type}'s"
    plt.figure(figsize=(12, 6))
    sns.barplot(x=boundary_counts.values, y=boundary_counts.index, palette="coolwarm")
    plt.title(title, fontsize=16)
    plt.xlabel(f"Number of {boundary_type}'s", fontsize=12)
    plt.ylabel("Batsmen", fontsize=12)
    plt.tight_layout()
    plt.show()

def highest_strike_rate(deliveries_df, min_balls=100):
    batsman_data = deliveries_df.groupby('batter').agg({'batsman_runs': 'sum', 'batter': 'size'}).rename(columns={'batter': 'balls_faced'})
    batsman_data = batsman_data[batsman_data['balls_faced'] >= min_balls]
    batsman_data['strike_rate'] = (batsman_data['batsman_runs'] / batsman_data['balls_faced']) * 100
    highest_sr = batsman_data.sort_values('strike_rate', ascending=False).head(10)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=highest_sr['strike_rate'], y=highest_sr.index, palette="Blues_r")
    plt.title("Batsmen with Highest Strike Rate", fontsize=16)
    plt.xlabel("Strike Rate", fontsize=12)
    plt.ylabel("Batsmen", fontsize=12)
    plt.tight_layout()
    plt.show()

def leading_wicket_takers(deliveries_df):
    wickets = deliveries_df[deliveries_df['dismissal_kind'].notnull()].groupby('bowler').size().sort_values(ascending=False).head(10)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=wickets.values, y=wickets.index, palette="coolwarm")
    plt.title("Leading Wicket-Takers of All Time", fontsize=16)
    plt.xlabel("Wickets Taken", fontsize=12)
    plt.ylabel("Bowlers", fontsize=12)
    plt.tight_layout()
    plt.show()

def stadium_with_most_matches(matches_df):
    stadium_counts = matches_df['venue'].value_counts()
    plt.figure(figsize=(12, 6))
    sns.barplot(x=stadium_counts.values, y=stadium_counts.index, palette="magma")
    plt.title("Stadiums Hosting the Most Matches", fontsize=16)
    plt.xlabel("Matches Hosted", fontsize=12)
    plt.ylabel("Stadiums", fontsize=12)
    plt.tight_layout()
    plt.show()

def most_mom_awards(matches_df):
    mom_counts = matches_df['player_of_match'].value_counts().head(10)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=mom_counts.values, y=mom_counts.index, palette="viridis")
    plt.title("Players with Most MOM Awards", fontsize=16)
    plt.xlabel("MOM Awards", fontsize=12)
    plt.ylabel("Players", fontsize=12)
    plt.tight_layout()
    plt.show()

def boundary_counts_per_season(deliveries_df, boundary_type):
    boundary_data = deliveries_df[deliveries_df['batsman_runs'] == boundary_type]
    counts = boundary_data.groupby('match_id')['match_id'].count()
    return counts.sum()

# Add remaining functions for boundary contributions, run rates, etc.

# Call functions for plots

matches_per_season(matches_df)
runs_per_season(deliveries_df, matches_df)
runs_per_match_per_season(matches_df)
most_frequent_umpires(matches_df)
most_tosses_won(matches_df)
toss_decisions(matches_df)
toss_decision_trend(matches_df)
toss_vs_match_winner(matches_df)
chasing_team_wins(matches_df)
teams_won_tournament(matches_df)
team_with_most_matches(matches_df)
team_with_most_wins(matches_df)
team_with_highest_win_percentage(matches_df)
lucky_venue_for_team(matches_df)
comparison_between_two_teams(deliveries_df, "Mumbai Indians", "Chennai Super Kings")
batsman_analysis(deliveries_df, "V Kohli")
teams_conceding_200_plus(deliveries_df)
highest_team_score(deliveries_df)
biggest_win_by_runs(matches_df)
most_balls_faced_batsmen(deliveries_df)
leading_run_scorers(deliveries_df)
most_boundaries(deliveries_df, boundary_type='4')
most_boundaries(deliveries_df, boundary_type='6')
highest_strike_rate(deliveries_df)
leading_wicket_takers(deliveries_df)
stadium_with_most_matches(matches_df)
most_mom_awards(matches_df)
```
